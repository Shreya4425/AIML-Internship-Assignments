```python
#Statistics
# 1. Mean
def mean(array):
    n = len(array)
    return np.sum(array)/n
```

```python
mean(np.array([1,2,3,4,5,6]))
```

```python
# 2. Variance
def variance(array):
    n = len(array)
    m = mean(array)
    return np.sum((array - m)**2)/n
```

```python
variance(np.array([1,2,3,4,5,6]))
```

```python
# 3. Standard Deviation
def std_deviation(array):
    return np.sqrt(variance(array))
```

```python
std_deviation(np.array([1,2,3,4,5,6]))
```

```python
# 4. Minimum & Maximum
def min_max(array):
    minimum = np.min(array)
    maximum = np.max(array)
    return minimum, maximum
```

```python
min_max(np.array([1,2,3,4,5,6]))
```

```python
# 5. Standard Scalar
def std_scalar(array):
    m = mean(array)
    std = std_deviation(array)
    return (array - m)/std
```

```python
std_scalar(np.array([1,2,3,4,5,6,7]))
```

```python
# 6. Sigmoid
def sigmoid(array):
    return 1/(1 + np.exp(-array))
```

```python
sigmoid(np.array([6,5,4,3,2,1]))
```

```python
# 7. Relu
def relu(array):
    return np.maximum(0,array)
```

```python
#relu(np.array([9,8,7,43,1,2,3,4,5,6]))
relu(np.array([-5, -2, 0, 3, 7]))
```
